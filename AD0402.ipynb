```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_excel(open('Sample.xlsx','rb'))
df
```

```python
model=IsolationForest(n_estimators=1000,max_samples='auto',contamination=float(0.2),max_features=1.0)
model.fit(df[['Cards']])
```

```python
df['cards']=model.decision_function(df[['Cards']])
df['anomaly']=model.predict(df[['Cards']])
df.head(30)
```
