```python
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
```

**Data description:**

*age*: age of person

*sex*: male/female

*bmi*:  body mass index

*children*: number of covered children/dependents

*smoker*: smoker/non-smoker

*region*: area of residence in the United States, northeast, southeast, southwest, northwest..

*charges*: insurance costs

```python
df = pd.read_csv('23.2 insurance.csv')
df
```

```python
def show_age_charges(df):
    plt.plot(df['age'], df.charges, 'ro')
    plt.xlabel('age')
    plt.ylabel('charges')
    plt.title('Dependence of the cost of insurance on age')
    plt.show()
```

```python
show_age_charges(df)
```

```python
df_new = df[(df.children==0) & (df.charges < 15000) & (df.sex == 'female')]
df_new.shape
```

```python
show_age_charges(df_new)
```

```python
train, test = train_test_split(df_new, test_size=0.2, random_state=42)
```

```python
train.head()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(train[['age']], train.charges)
```

MAE (mean absolute error)

is calculated as the average of the absolute differences between the target values and predictors.Снимок экрана 2022-04-06 в 13.31.32.png

Example:

true = [0, 100, 200]

pred = [0, 50, 250]

MAE = (0 + 50 + 50)/3 = 33,3

```python
true = [0, 100, 200]
pred = [0, 50, 250]
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(true, pred)
```

```python
pred = lr.predict(test[['age']])
mean_absolute_error(test.charges, pred)
```

```python
indx = df_new[(df_new.charges >= 6000) & (df_new.age<50)].index
print(df_new.shape)
df_new = df_new.drop(indx)
print(df_new.shape)
```

```python
train, test = train_test_split(df_new, test_size=0.2, random_state=42)

lr.fit(train[['age']], train.charges)
pred = lr.predict(test[['age']])
mean_absolute_error(test.charges, pred)
```
